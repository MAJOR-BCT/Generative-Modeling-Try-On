```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

```python
# Generate two random images of size 16x16
random_image_1 = np.random.randint(0, 255, (16, 16)).astype(np.uint8)
random_image_2 = np.random.randint(0, 255, (16, 16)).astype(np.uint8)

# Apply Sobel operations to calculate gradients
sobel_x_img_1 = cv2.Sobel(random_image_1, cv2.CV_64F, 1, 0, ksize=3)
sobel_y_img_1 = cv2.Sobel(random_image_1, cv2.CV_64F, 0, 1, ksize=3)

sobel_x_img_2 = cv2.Sobel(random_image_2, cv2.CV_64F, 1, 0, ksize=3)
sobel_y_img_2 = cv2.Sobel(random_image_2, cv2.CV_64F, 0, 1, ksize=3)

# Display the original and Sobel gradient images
plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.title('Random Image 1')
plt.imshow(random_image_1, cmap='gray')
plt.axis('off')

plt.subplot(222)
plt.title('Sobel X - Random Image 1')
plt.imshow(cv2.convertScaleAbs(sobel_x_img_1), cmap='gray')
plt.axis('off')

plt.subplot(223)
plt.title('Random Image 2')
plt.imshow(random_image_2, cmap='gray')
plt.axis('off')

plt.subplot(224)
plt.title('Sobel X - Random Image 2')
plt.imshow(cv2.convertScaleAbs(sobel_x_img_2), cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def create_digit_image(digit):
    img = np.zeros((32, 32), dtype=np.uint8)
    if digit == 0:
        # Draw digit 0
        cv2.circle(img, (16, 16), 12, (255), 1)
    elif digit == 1:
        # Draw digit 1
        cv2.line(img, (4, 3), (4, 23), (255), 1)
    return img

# Generate images of digits 0 and 1
digit_0 = create_digit_image(0)
digit_1 = create_digit_image(1)

# Apply Sobel operations to calculate gradients
sobel_x_digit_0 = cv2.Sobel(digit_0, cv2.CV_64F, 1, 0, ksize=3)
sobel_y_digit_0 = cv2.Sobel(digit_0, cv2.CV_64F, 0, 1, ksize=3)

sobel_x_digit_1 = cv2.Sobel(digit_1, cv2.CV_64F, 1, 0, ksize=3)
sobel_y_digit_1 = cv2.Sobel(digit_1, cv2.CV_64F, 0, 1, ksize=3)

# Display the original digits and Sobel gradient images
plt.figure(figsize=(10, 8))

plt.subplot(241)
plt.title('Digit 0')
plt.imshow(digit_0, cmap='gray')
plt.axis('off')

plt.subplot(242)
plt.title('Sobel X - Digit 0')
plt.imshow(cv2.convertScaleAbs(sobel_x_digit_0), cmap='gray')
plt.axis('off')

plt.subplot(243)
plt.title('Sobel Y - Digit 0')
plt.imshow(cv2.convertScaleAbs(sobel_y_digit_0), cmap='gray')
plt.axis('off')

plt.subplot(245)
plt.title('Digit 1')
plt.imshow(digit_1, cmap='gray')
plt.axis('off')

plt.subplot(246)
plt.title('Sobel X - Digit 1')
plt.imshow(cv2.convertScaleAbs(sobel_x_digit_1), cmap='gray')
plt.axis('off')

plt.subplot(247)
plt.title('Sobel Y - Digit 1')
plt.imshow(cv2.convertScaleAbs(sobel_y_digit_1), cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
